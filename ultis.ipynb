```python
import os
# root_dir = "generated_dataset/scaled_data"
# root_dir = "generated_dataset/rotated_data/"
# root_dir = "generated_dataset/noised_data/"
root_dir = "generated_dataset/noised_data/noised_original_data"
# root_dir = "generated_dataset/combine_3_with_others_data/"
```

```python
# List of folders
# folders = ["scaled_biased_rate_0.3_data", "scaled_biased_rate_0.35_data", "scaled_original_data"]
# folders = ["rotated_biased_rate_0.3_data", "rotated_biased_rate_0.35_data", "rotated_original_data"]
# folders = ["noised_biased_rate_0.3_data", "noised_biased_rate_0.35_data"]
folders = ["gaussian_noised_data", "multiplicative_noised_data"]
# folders = ["biased_rate_0.3_data", "biased_rate_0.35_data", "original_data" ]

# Iterate through each folder
for folder in folders:
    images_folder = os.path.join(root_dir, folder, "images")
    labels_folder = os.path.join(root_dir, folder, "labels")
    print(images_folder)
    # Iterate through images
    for image_file in os.listdir(images_folder):
        print(image_file)
        if image_file.endswith(".jpg"):
            old_path = os.path.join(images_folder, image_file)
            new_name = folder + "_" + image_file
            new_path = os.path.join(images_folder, new_name)
            os.rename(old_path, new_path)
    
    # Iterate through label files
    for label_file in os.listdir(labels_folder):
        if label_file.endswith(".txt"):
            old_path = os.path.join(labels_folder, label_file)
            new_name = folder + "_" + label_file
            new_path = os.path.join(labels_folder, new_name)
            os.rename(old_path, new_path)

print("File renaming completed successfully.")
```
